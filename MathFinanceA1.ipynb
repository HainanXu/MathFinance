```python
import numpy as np
import math

def bootstrap_spot_rate(bonds):
    """
    Calculate the spot rates for a series of bonds using the bootstrap method.

    Parameters:
    - bonds: A list of tuples/lists where each tuple/list contains the bond's attributes in the order:
             (index, time to maturity, coupon rate, present value).
             The coupon rate is assumed to be annual, and payments are semi-annual.

    Returns:
    - s: An array of calculated spot rates for each bond.
    """
    s = np.zeros(len(bonds))  # Initialize an array for spot rates with the same length as the bonds list
    for i in range(len(bonds)):
        ttm = bonds[i][1]
        pv = bonds[i][3]
        coupon = bonds[i][2]
        if i == 0:
            # for the first bond, adjust future value for semi-annual coupon payments and the accrued data
            fv = 100 + (coupon/2) * (ttm/0.5)
            s[i] = (fv/pv)**(1/ttm) - 1
        else:
            # for subsequent bonds, calculate spot rate using cash flows adjusted for previous spot rates
            cf = np.zeros(i + 1)
            cf[:] = (coupon/2)  # semi-annual coupon payments
            cf[-1] += 100  # add the principal to the last cash flow

            npv2 = bonds[i][3]
            for i2 in range(len(cf) - 1):
                npv2 -= cf[i2] / (1 + s[i2])**bonds[i2][1]
            s[i] = (cf[-1] / npv2)**(1/bonds[i][1]) - 1
    return s
```

```python
bonds_list= list()
spot_rates_dict={}
date_columns= Bond.columns[5:16]
for date in date_columns:
  formatted_bonds = [(row['ISIN'], row['ttm'], row['coupon'], row[date]) for index, row in Bond.iterrows()]
  bonds = [list(bond) for bond in formatted_bonds]
  spot_rates=bootstrap_spot_rate(bonds)
  bonds_list.append(bonds)
  spot_rates_dict[date]=spot_rates

spot_rates_dict
```

```python
def calculate_forward_rate(spot_rates, T1, T2):
    """
    Calculate the forward rate from time T1 to T2 given an array of spot rates.

    Parameters:
    - spot_rates: An array of spot rates for each time period.
    - T1: The starting time period for the forward rate.
    - T2: The ending time period for the forward rate.

    Returns:
    - The forward rate from T1 to T2.
    """
    # Adjusting for Python's 0-based indexing
    T1_adj = T1 - 1
    T2_adj = T2 - 1

    forward_rate = (((1 + spot_rates[T2_adj])**(2*T2) / (1 + spot_rates[T1_adj])**(2*T1))**(1 / (2*(T2 - T1)))) - 1
    return forward_rate
```

```python
forward_rate_dict = {}
df = pd.DataFrame(spot_rates_dict)

for col in df:
    forward_rates = np.zeros(4)
    for i in range(2, 6):
        spot_rates = df[col].values
        forward_rate = calculate_forward_rate(spot_rates, 1, i)
        forward_rates[i-2] = forward_rate

    forward_rate_dict[col] = forward_rates


df2=pd.DataFrame(forward_rate_dict)
df2
```
